```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ticker = "NVDA"
df = yf.download(ticker, start="2015-01-01", end="2024-12-31")
df.head()
```

```python

df.isnull().sum()


```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Close'], label='Closing Price')
plt.title(f"{ticker} Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()



```

```python


df['Daily_Return'] = df['Close'].pct_change()
df = df.dropna()
df.head()

```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Daily_Return'], color='orange')
plt.title(f"{ticker} Daily Returns")
plt.xlabel("Date")
plt.ylabel("Return")
plt.show()
```

```python
mean_return = df['Daily_Return'].mean()
variance_return = df['Daily_Return'].var()
volatility = df['Daily_Return'].std()

mean_return, variance_return, volatility
```

## Statistical Interpretation

- **Mean Return**: Average daily change in price  
- **Variance**: Spread of returns  
- **Volatility (Standard Deviation)**: Market risk indicator  

Higher volatility means higher uncertainty and risk.
